```
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```
a = cv.imread("Mark_Zukerberg.png")
plt.imshow(cv.cvtColor(a, cv.COLOR_BGR2RGB))
```

```
b = cv.cvtColor(a, cv.COLOR_BGR2GRAY)
plt.imshow(b, cmap='gray')
```

```
def convolution(image, mask, stride=1):
    rows, cols = image.shape
    k_rows, k_cols = mask.shape
    out_rows = (rows - k_rows) // stride + 1
    out_cols = (cols - k_cols) // stride + 1
    output = np.zeros((out_rows, out_cols), dtype=np.float64)
    for i in range(out_rows):
        for j in range(out_cols):
            roi = image[i*stride:i*stride+k_rows, j*stride:j*stride+k_cols]
            output[i, j] = np.sum(roi * mask)
    return output
```

```
b.shape
```

```
scaled = b.astype(np.float64) / 255.0
```

```
sobel_x = np.array([[-1, 0, 1],
                     [-2, 0, 2],
                     [-1, 0, 1]], dtype=np.float64)

conv_output1 = convolution(scaled, sobel_x, 1)
conv_output2 = convolution(scaled, sobel_x, 2)
```

```
print(conv_output1.shape)
print(conv_output2.shape)
```

```
relu_output1 = np.maximum(0, conv_output1)
relu_output2 = np.maximum(0, conv_output2)
```

```
def max_pooling(image, pool_size=2):
    h, w = image.shape
    ph, pw = h // pool_size * pool_size, w // pool_size * pool_size
    image = image[:ph, :pw]
    pooled = image.reshape(ph // pool_size, pool_size, pw // pool_size, pool_size).max(axis=(1, 3))
    return pooled

pooled_output1 = max_pooling(relu_output1, pool_size=2)
pooled_output2 = max_pooling(relu_output2, pool_size=2)
```

```
flattened1 = pooled_output1.flatten()
flattened2 = pooled_output2.flatten()
```

```
fig, axes = plt.subplots(1, 4, figsize=(16, 4))
axes[0].imshow(scaled, cmap='gray')
axes[0].set_title('Scaled (0-1)')
axes[1].imshow(conv_output1, cmap='gray')
axes[1].set_title('After Convolution')
axes[2].imshow(relu_output1, cmap='gray')
axes[2].set_title('After ReLU')
axes[3].imshow(pooled_output1, cmap='gray')
axes[3].set_title('After Max Pooling')
for ax in axes:
    ax.axis('off')
plt.tight_layout()
plt.show()

print(f"Flattened shape: {flattened1.shape}")
```
