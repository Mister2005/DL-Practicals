```python
import numpy as np

def unipolar_step(net):
    if net >= 0:
        return 1
    elif net < 0:
        return 0

def bipolar_step(net):
    if net > 0:
        return 1
    elif net == 0:
        return 0
    elif net < 0:
        return -1
```

```python
def perceptron(data, targets, w, b, lr, act_func):
    epoch = 0
    while True:
        epoch += 1
        print(f"Epoch {epoch}")
        del_w = False
        for i, (x,d) in enumerate(zip(data, targets)):
            net = np.dot(w, x) + b
            o = act_func(net)
            e = d - o
            if e != 0:
                del_w = True
                new_w = lr * e * x
                new_b = lr * e
                w = w + new_w
                b = b + new_b
            print(f"  Step {i + 1}: Input={x}, Net={net}, Target={d}, o={o}, Error={e}, Weights={w}, Bias={b}")
        if not del_w:
            print(f"Converged at Epoch {epoch}")
            break
    return w, b
```

# AND

```python
print("Bipolar Step Function")
bi_input = np.array([[-1, -1], [-1, 1], [1, -1], [1, 1]])
bi_targets = np.array([-1, -1, -1, 1])
bi_w = np.array([0, 0])
bi_b = 0
bi_lr = 1
final_bi_w, final_bi_b = perceptron(bi_input, bi_targets, bi_w, bi_b, bi_lr, bipolar_step)
print(f"Final Bipolar Weights: {final_bi_w}, Final Bias: {final_bi_b}\n")

print("Unipolar Step Function")
uni_input = np.array([[0, 0], [0, 1], [1, 0], [1, 1]]) #uni_input = np.array([[-1, -1], [-1, 1], [1, -1], [1, 1]])
uni_targets = np.array([0, 0, 0, 1])
uni_w = np.array([0, 0])
uni_b = 0
uni_lr = 1
final_uni_w, final_uni_b = perceptron(uni_input, uni_targets, uni_w, uni_b, uni_lr, unipolar_step)
print(f"Final Unipolar Weights: {final_uni_w}, Final Bias: {final_uni_b}")
```

# Static Sum

```python
print("Static Sum")
uni_input = np.array([[0, 0], [1, 0], [0, 1], [1, 1]])

w_or, b_or = np.array([1.0, 1.0]), -0.5
w_nand, b_nand = np.array([-1.0, -1.0]), 1.5
w_and, b_and = np.array([1.0, 1.0]), -1.5

for x in uni_input:
    out_or = unipolar_step(np.dot(w_or, x) + b_or)
    out_nand = unipolar_step(np.dot(w_nand, x) + b_nand)
    hidden_out = np.array([out_or, out_nand])
    sum_out = unipolar_step(np.dot(w_and, hidden_out) + b_and)
    print(f"Input: {x}, Sum: {sum_out}")
```
