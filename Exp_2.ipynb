```python
import numpy as np

def uni_sig(x):
    return 1 / (1 + np.exp(-x))

def uni_sig_der(O):
    return O * (1 - O)

def bi_sig(x):
    return (2 / (1 + np.exp(-x))) - 1

def bi_sig_der(O):
    return (1 - np.square(O)) / 2
```

```python
def forward_pass(x, w1, b1, w2, b2, activation_fn):
    net_h = np.dot(x, w1) + b1
    o_h = activation_fn(net_h)
    
    net_o = np.dot(o_h, w2) + b2
    o_o = activation_fn(net_o)
    
    return net_h, o_h, net_o, o_o

def backward_pass(x, d, net_h, o_h, net_o, o_o, w2, activation_der_fn):
    m = x.shape[0]
    
    dnet_o = (o_o - d) * activation_der_fn(o_o)
    dw2 = np.dot(o_h.T, dnet_o) / m
    db2 = np.sum(dnet_o, axis=0, keepdims=True) / m
    
    do_h = np.dot(dnet_o, w2.T)
    dnet_h = do_h * activation_der_fn(o_h)
    dw1 = np.dot(x.T, dnet_h) / m
    db1 = np.sum(dnet_h, axis=0, keepdims=True) / m
    
    return dw1, db1, dw2, db2
```

# Unipolar Activation Function

```python
lr = 1

x = np.array([[1, 0]])
d_target = np.array([[1]])

w1 = np.array([[0.5, 0.8], 
               [-0.3, 0.2]])
b1 = np.array([[0.2, -0.4]])

w2 = np.array([[0.6], 
               [-0.5]])
b2 = np.array([[0.3]])

epoch = 0
mse = 1

while epoch < 500 and mse > 0:
    net_h, out_h, net_o, out_o = forward_pass(x, w1, b1, w2, b2, uni_sig)
    
    mse = 0.5 * (d_target[0][0] - out_o[0][0]) ** 2
    
    dw1, db1, dw2, db2 = backward_pass(x, d_target, net_h, out_h, net_o, out_o, w2, uni_sig_der)
    
    w1 = w1 - lr * dw1
    b1 = b1 - lr * db1
    
    w2 = w2 - lr * dw2
    b2 = b2 - lr * db2
    
    epoch += 1
    
    if epoch % 100 == 0 or epoch == 1:
        print(f"Epoch {epoch}: MSE = {mse:.10f}, Output = {out_o[0][0]:.6f}")

print("\n" + "=" * 50)
print("TRAINING COMPLETE")
print("=" * 50)
print(f"Total Epochs: {epoch}")
print(f"Final MSE: {mse:.10f}")
print(f"Final Output: {out_o[0][0]:.6f}")
print(f"\nFinal Weights:")
print(f"w1: {w1[0][0]:.4f}, w2: {w1[1][0]:.4f}, w3: {w1[0][1]:.4f}, w4: {w1[1][1]:.4f}, w5: {w2[0][0]:.4f}, w6: {w2[1][0]:.4f}")
print(f"\nFinal Biases:")
print(f"b1: {b1[0][0]:.4f}, b2: {b1[0][1]:.4f}, b3: {b2[0][0]:.4f}")
```

# Bipolar Sigmoid Function

# XOR

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = np.array([[0], [1], [1], [0]])

W1 = np.array([[0.5, -0.3], [0.4, 0.2]])
W2 = np.array([[0.1], [0.3]])
b1 = np.array([[0.3, -0.2]])
b2 = np.array([[0.5]])

lr = 1
effective_lr = lr * X.shape[0]

for epoch in range(5000):
    net_h, H, net_o, O = forward_pass(X, W1, b1, W2, b2, uni_sig)
    
    dW1, db1, dW2, db2 = backward_pass(X, Y, net_h, H, net_o, O, W2, uni_sig_der)
    
    W2 -= effective_lr * dW2
    b2 -= effective_lr * db2
    W1 -= effective_lr * dW1
    b1 -= effective_lr * db1

print("Outputs:", O.flatten())
```
